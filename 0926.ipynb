```python
import pandas as pd
from itertools import combinations

# CSV 파일 읽기
file_path = r'C:\Users\SNUH\Documents\github\CDM_hemodialysis\distinct.csv'
df = pd.read_csv(file_path)

# 결과를 저장할 딕셔너리 초기화
results = {}

# 각 환자의 고유 세트를 저장할 딕셔너리
patient_sets = {}

# 각 환자의 고유 세트 생성
for pt_id in df['PT_ID'].unique():
    patient_data = df[df['PT_ID'] == pt_id]
    patient_sets[pt_id] = set(tuple(row) for row in patient_data[['BFR', 'Pre', 'Replace', 'Dialysate']].itertuples(index=False))

# 환자 쌍 비교
overlapping_pairs = []
for (pt1, sets1), (pt2, sets2) in combinations(patient_sets.items(), 2):
    if sets1.intersection(sets2):
        overlapping_pairs.append((pt1, pt2))

# 결과 출력
if overlapping_pairs:
    print("다음 환자 쌍 사이에 중복되는 세트가 있습니다:")
    for pt1, pt2 in overlapping_pairs:
        print(f"  {pt1} 와 {pt2}")
    print("\n중복되는 세트:")
    for pt1, pt2 in overlapping_pairs:
        common_sets = patient_sets[pt1].intersection(patient_sets[pt2])
        for common_set in common_sets:
            print(f"  환자 {pt1} 와 {pt2}: {common_set}")
else:
    print("모든 환자의 세트가 고유합니다. 중복되는 세트가 없습니다.")

# 각 환자의 고유 세트 수 출력
print("\n각 환자의 고유 세트 수:")
for pt_id, sets in patient_sets.items():
    print(f"  환자 {pt_id}: {len(sets)} 개의 고유 세트")
```
